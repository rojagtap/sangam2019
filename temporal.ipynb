```
# filter device_id
# remove all {"success": 1}
df = df[df.device_id != "{\"success\":1"]
df = df[df.device_id != "S81"]

# Lat and Long of IIT Madras ISRC
df[df.device_id == "M7"].latitude = 12.9917081
df[df.device_id == "M7"].longitude = 80.2320804

# Given in PS
df[df.device_id == "S9"].latitude = 13.088279
df[df.device_id == "S9"].longitude = 80.181568

df = df[~df.device_id.isnull()]
```

```
df.describe(include="all")
```

```
# remove redundant columns
df.drop(['nso', 'ewo'], axis=1, inplace=True)
```

```
df.loc[df.timestamp == "0000-00-00 00:00:00", 'timestamp'] = df.loc[df.timestamp == "0000-00-00 00:00:00", 'svrtime']
```

```
df.timestamp = pd.to_datetime(df.timestamp)
df.svrtime = pd.to_datetime(df.svrtime)
```

```
# replace 0 dates with server time as the difference between timestamp and
# server time is supposed to be very less and the time in seconds is not available

df.loc[abs((df.svrtime - df.timestamp)/np.timedelta64(1, 'M')) > 1, 'timestamp'] = df.loc[abs((df.svrtime - df.timestamp)/np.timedelta64(1, 'M')) > 1, 'svrtime']
```

```
#split into chennai and gurgaon
sensor_gurgaon = ["M1", "M2", "M3", "M4", "S1", "S3", "S4", "S5"]
sensor_chennai = ["M5", "M6", "S6", "S7", "S8", "S9", "S10", "M7"]
chennai = df[df.device_id.isin(sensor_chennai)]
gurgaon = df[df.device_id.isin(sensor_gurgaon)]
```

```
# removal of lat long outliers

#df.latitude = df.latitude.apply(lambda x: x if (11 < x < 15) or (26 < x < 30) else np.nan)
#df.longitude = df.longitude.apply(lambda x: x if (75 < x < 82) else np.nan)
chennai.latitude = chennai.latitude.apply(lambda x: x if 12.99 <= x <= 13.14 else np.nan)
gurgaon.latitude = gurgaon.latitude.apply(lambda x: x if 28.42 <= x < 28.48 else np.nan)

chennai.longitude = chennai.longitude.apply(lambda x: x if 80.139 <= x <= 80.233 else np.nan)
gurgaon.longitude = gurgaon.longitude.apply(lambda x: x if 77.028 <= x <= 77.090 else np.nan)
```

### predicting in form, [yd0, yd1,...,yd6] where y is the AQI predicted for the next week

```
def model_x_train(df):
    x_temp = []
    for i in range(df.shape[0] - 2):
        for j in range(df.shape[1]):
            x_temp.append(np.concatenate([df[i][j:], df[i + 1][:j]], axis=0))
    return x_temp
```

```
def model_y_train(df): 
    y_temp = []
    for i in range(1, df.shape[0] - 1):
        for j in range(df.shape[1]):
            y_temp.append(np.concatenate((df[i][j:], df[i + 1][:j])))
    return y_temp
```

```
x_train_chennai = np.array(model_x_train(x_train_chennai))
y_train_chennai = np.array(model_y_train(y_train_chennai))
```
